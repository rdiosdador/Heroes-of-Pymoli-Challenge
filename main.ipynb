```python
import pandas as pd
```

```python
filePath = "Resources/purchase_data.csv"

purchase_data = pd.read_csv(filePath)

purchase_data.head()
```

```python
#Player Count

#Total Number of Players
```

```python
 
playersCount = len(purchase_data['SN'].value_counts())

#Create a new data frame to store playersCount

totalNumPla = pd.DataFrame([playersCount], columns= ['Total Players'])

totalNumPla

```

```python
#Purchasing Analysis (Total)

#Number of Unique Items
#Average Purchase Price
#Total Number of Purchases
#Total Revenue
```

```python
#Number of unique items

uniqueItems = len(purchase_data["Item Name"].unique())

uniqueItems
```

```python
#Averge of price 

averagePrice = purchase_data['Price'].mean()

averagePrice
```

```python
#Number purchases

numPurchases = len(purchase_data['Purchase ID'])

numPurchases
```

```python
#Total revenue

totalRevenue = purchase_data['Price'].sum()

totalRevenue
```

```python
#Create a new data frame to store the Purchasing Analysis


purchasingAnalysis = pd.DataFrame({'Number of Unique Items':[uniqueItems],
                                   'Average Price':[averagePrice],
                                   'Number of purchases':[numPurchases],
                                   'Total Revenue':[totalRevenue]}
                                  )

# Format dataframe columns
#format_mapping={'Currency': '${:,.2f}', 'Int': '{:,.0f}', 'Rate': '{:.2f}%'}

purchasingAnalysis["Total Revenue"] = purchasingAnalysis["Total Revenue"].map("${:,.2f}".format)
purchasingAnalysis["Average Price"] = purchasingAnalysis["Average Price"].map("{:,.2f}".format)

purchasingAnalysis
```

```python
#Gender Demographics

#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

```

```python
#Create a new data frame to store only the columns 'SN' and 'Gender' (purchase_data2 contains duplicate values)

purchase_data2 = purchase_data[["SN","Gender"]]

purchase_data2.groupby(['Gender']).count()

```

```python
#Create a new data frame to store data (No duplicate values in 'SN')

purchase_data3 = purchase_data2.drop_duplicates (["SN"])

purchase_data_nd = purchase_data3.groupby(['Gender']).count()

purchase_data_nd
```

```python
# Total Count by gender

male = len(purchase_data3.loc[purchase_data3["Gender"]=="Male", :])
female = len(purchase_data3.loc[purchase_data3['Gender']=='Female',:])
other = len(purchase_data3.loc[purchase_data3['Gender']=='Other / Non-Disclosed',:])
totalCount = len(purchase_data3)
```

```python
# Percent of Players by gender

percentMale = (male/totalCount) * 100
percentFemale = (female/totalCount) * 100
percentOther = (other/totalCount) * 100
```

```python
#Create a new data frame to store the gender Analysis

genderDemographics = pd.DataFrame([[percentFemale, female],
                              [percentMale, male],
                              [percentOther, other]],
                              index=["Female","Male", "Other / Non-Disclosed"],
                              columns=["Percentage of Players", "Total Count"])

# Format dataframe columns

genderDemographics["Percentage of Players"] = genderDemographics["Percentage of Players"].map("{:.2f}%".format)

genderDemographics
```

```python
#Purchasing Analysis (Gender)

#The below each broken by gender

#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Average Purchase Total per Person by Gender
```

```python
#Purchase Count by gender

malePurch = purchase_data[purchase_data["Gender"] == "Male"]["Purchase ID"].count()
femalePurch = purchase_data[purchase_data["Gender"] == "Female"]["Purchase ID"].count()
otherPurch = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]["Purchase ID"].count()

#Average Purchase Price

malePurchAvg = purchase_data[purchase_data["Gender"] == "Male"]['Price'].mean()
femPurchAvg = purchase_data[purchase_data["Gender"] == "Female"]['Price'].mean()
othPurchAvg = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]['Price'].mean()

#Total Purchase Value

malePurchTot = purchase_data[purchase_data["Gender"] == "Male"]['Price'].sum()
femPurchTot = purchase_data[purchase_data["Gender"] == "Female"]['Price'].sum()
othPurchTot = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]['Price'].sum()

#Average Purchase Total per Person by Gender

malePerGen = malePurchTot/male
femPerGen = femPurchTot/female
othPerGen = othPurchTot/other

```

```python
#Create new dataframe to store Purchasing Analysis (Gender)

purchasingGender = pd.DataFrame([[malePurch,malePurchAvg,malePurchTot,malePerGen],
                              [femalePurch,femPurchAvg,femPurchTot,femPerGen],
                              [otherPurch,othPurchAvg,othPurchTot,othPerGen]],
                              index=["Female","Male", "Other / Non-Disclosed"],
                              columns=["Purchase Count", "Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"])


# Format dataframe columns
#format_mapping={'Currency': '${:,.2f}', 'Int': '{:,.0f}', 'Rate': '{:.2f}%'}

purchasingGender["Average Purchase Price"] = purchasingGender["Average Purchase Price"].map("{:,.2f}".format)
purchasingGender["Total Purchase Value"] = purchasingGender["Total Purchase Value"].map("{:,.2f}".format)
purchasingGender["Avg Total Purchase per Person"] = purchasingGender["Avg Total Purchase per Person"].map("{:,.2f}".format)

purchasingAnalysis


purchasingGender
```

```python
#Age Demographics

#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)

#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Average Purchase Total per Person by Age Group
```

```python

bins = [0, 9.5, 14.5, 19.5, 24.5, 29.5, 34.5, 39.5, 1000]

binLabels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age group"] = pd.cut(purchase_data["Age"], bins, labels = binLabels)

ageGrouped = purchase_data.groupby("Age group")

uniqueMembers = ageGrouped["SN"].nunique()

ageDemographics = pd.DataFrame({"Total Counts": ageGrouped["SN"].nunique(),
                           "Percentage of Players": 100*uniqueMembers/playersCount})

#format_mapping={'Currency': '${:,.2f}', 'Int': '{:,.0f}', 'Rate': '{:.2f}%'}

ageDemographics["Percentage of Players"] = ageDemographics["Percentage of Players"].map('{:,.2f}%'.format)

ageDemographics
```

```python
#Purchasing Analysis (Age)

#Bin the purchase_data data frame by age
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame
```

```python
purchaseCount = ageGrouped["Age"].count()
avgPurchase = ageGrouped["Price"].mean()
totalPurchaseValue = ageGrouped["Price"].sum()
```

```python
purchAnalysAge = pd.DataFrame({"Purchase count": purchaseCount,
                                       "Average purchase price": avgPurchase,
                                       "Total purchase value": totalPurchaseValue,
                                        "Avg Total Purchase per Person": totalPurchaseValue/uniqueMembers})

#format_mapping={'Currency': '${:,.2f}', 'Int': '{:,.0f}', 'Rate': '{:.2f}%'}

purchAnalysAge["Average purchase price"] = purchAnalysAge["Average purchase price"].map('${:,.2f}'.format)
purchAnalysAge["Total purchase value"] = purchAnalysAge["Total purchase value"].map('${:,.2f}'.format)
purchAnalysAge["Avg Total Purchase per Person"] = purchAnalysAge["Avg Total Purchase per Person"].map('${:,.2f}'.format)

purchAnalysAge
```

```python
#Top Spenders

#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

#SN
#Purchase Count
#Average Purchase Price
#Total Purchase Value
```

```python
spendersGrouped = purchase_data.groupby("SN")

purchaseCount2 = spendersGrouped["Age"].count()
avgPurchase2 = spendersGrouped["Price"].mean()
totalPurchaseValue2 = spendersGrouped["Price"].sum()
```

```python
topSpenders = pd.DataFrame({"Purchase count": purchaseCount2,
                                       "Average purchase price": avgPurchase2,
                                       "Total purchase value": totalPurchaseValue2})

topSpenders = topSpenders.sort_values("Total purchase value", ascending = False)

#format_mapping={'Currency': '${:,.2f}', 'Int': '{:,.0f}', 'Rate': '{:.2f}%'}

topSpenders["Average purchase price"] = topSpenders["Average purchase price"].map('${:,.2f}'.format)
topSpenders["Total purchase value"] = topSpenders["Total purchase value"].map('${:,.2f}'.format)

topSpenders.head(5)
```

```python
#Most Popular Items

#Identify the 5 most popular items by purchase count, then list (in a table):

#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
```

```python
itemsGrouped = purchase_data.groupby(["Item ID", "Item Name"])

purchaseCount3 = itemsGrouped["Purchase ID"].count()
totalPurchaseValue3 = itemsGrouped["Price"].sum()
itemPrice = itemsGrouped["Price"].max()
```

```python
topItems = pd.DataFrame({"Purchase count": purchaseCount3,
                                       "Item price": itemPrice,
                                       "Total purchase value": totalPurchaseValue3})

# Sort the data frame by "Purchase count" descending
topItems = topItems.sort_values("Purchase count", ascending = False)

#format_mapping={'Currency': '${:,.2f}', 'Int': '{:,.0f}', 'Rate': '{:.2f}%'}
topItems["Item price"] = topItems["Item price"].map('${:,.2f}'.format)
topItems["Total purchase value"] = topItems["Total purchase value"].map('${:,.2f}'.format)

topItems.head(5)
```

```python
#Most Profitable Items

#Identify the 5 most profitable items by total purchase value, then list (in a table):

#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
```

```python
mostProfitItems = pd.DataFrame({"Purchase count": purchaseCount3,
                                       "Item price": itemPrice,
                                       "Total purchase value": totalPurchaseValue3})

# Sort the data frame by "Total Purchase Value" descending

mostProfitItems = mostProfitItems.sort_values("Total purchase value", ascending = False)

#format_mapping={'Currency': '${:,.2f}', 'Int': '{:,.0f}', 'Rate': '{:.2f}%'}

mostProfitItems["Item price"] = mostProfitItems["Item price"].map('${:,.2f}'.format)
mostProfitItems["Total purchase value"] = mostProfitItems["Total purchase value"].map('${:,.2f}'.format)

mostProfitItems.head(5)
```
